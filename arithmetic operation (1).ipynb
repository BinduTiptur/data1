```python
#arithmetic operations
#multiplication
import cv2
import matplotlib.pyplot as plt
imge1=plt.imread('rose.jpg')
imge2=plt.imread('ros.jpg')
plt.imshow(imge1)
plt.show()
plt.imshow(imge2)
plt.show()
addimg=imge1*imge2
plt.imshow(addimg)
plt.show()
```

```python
#addition
import cv2
import matplotlib.pyplot as plt
imge1=plt.imread('rose.jpg')
imge2=plt.imread('ros.jpg')
plt.imshow(imge1)
plt.show()
plt.imshow(imge2)
plt.show()
addimg=imge1+imge2
plt.imshow(addimg)
plt.show()
```

```python
#subtraction
import cv2
import matplotlib.pyplot as plt
imge1=plt.imread('rose.jpg')
imge2=plt.imread('ros.jpg')
plt.imshow(imge1)
plt.show()
plt.imshow(imge2)
plt.show()
addimg=imge1-imge2
plt.imshow(addimg)
plt.show()
```

```python
#division
import cv2
import matplotlib.pyplot as plt
imge1=plt.imread('rose.jpg')
imge2=plt.imread('ros.jpg')
plt.imshow(imge1)
plt.show()
plt.imshow(imge2)
plt.show()
addimg=imge1/imge2
plt.imshow(addimg)
plt.show()
```
