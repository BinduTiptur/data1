```python
#original image
import cv2
import matplotlib.pyplot as plt
from PIL import Image
im = Image.open("rose.jpg")
plt.imshow(im)
plt.show()
```

```python
#upsampling
import cv2
import matplotlib.pyplot as plt
from PIL import Image
im = Image.open("rose.jpg")
im = im.resize((im.width*5, im.height*5), Image.NEAREST)
plt.figure(figsize=(10,10))
plt.title("upsampling")
plt.imshow(im)
plt.show()
```

```python
#downsampling
im = Image.open("rose.jpg")
im = im.resize((im.width//5, im.height//5))
plt.figure(figsize=(15,10))
plt.title("down sampling")
plt.imshow(im)
plt.show()
```
